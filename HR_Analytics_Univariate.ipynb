```python
import numpy as np
import pandas as pd
import seaborn as sns
from Univariate import Univariate
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv("HR_Analytics.csv")
```

```python
dataset
```

```python
dataset=dataset.drop("EmployeeCount",axis=1) # droping the column since count is always 1
dataset=dataset.drop("StandardHours",axis=1) #droping the column since StandardHours is always 80
```

```python
dataset.info()
```

```python
quan,qual = Univariate.quanqual(dataset)
```

```python
dataset.describe()
```

```python
descriptive = Univariate.unidesc(dataset,quan)
```

```python
lesser,greater = Univariate.findingoutlier(quan,descriptive)
print(lesser)
print(greater)
```

```python
dataset = Univariate.replacingoutlier(dataset,descriptive,lesser,greater)
```

```python
descriptive = Univariate.unidesc(dataset,quan)
descriptive
```

```python
lesser,greater = Univariate.findingoutlier(quan,descriptive)
print(lesser)
print(greater)
```
