```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv("HR_Analytics.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset=dataset.drop("EmployeeCount",axis=1) # droping the column since count is always 1
dataset=dataset.drop("StandardHours",axis=1) #droping the column since StandardHours is always 80
```

```python
def quanqual(dataset):
    qual = []
    quan = []
    for columnName in dataset.columns:
        if (dataset[columnName].dtype == 'O'):
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan,qual
```

```python
quan,qual = quanqual(dataset)
```

```python
quan_df = dataset[quan]
```

```python
qual_df = dataset[qual]
```

```python
dataset.describe()
```

```python
preprocessed = pd.concat([quan_df,qual_df],axis = 1)
preprocessed
```

```python
preprocessed=pd.get_dummies(preprocessed,drop_first=True).astype(int)
```

```python
preprocessed.to_csv("Prep_HR_Analytics.csv",index = False)
```
